```python
# lets make all  the duplicate values and make it upper case

def to_uppercase(column):
  return column.upper()
```

```python
df['ship-state']=df['ship-state'].apply(to_uppercase)
```

```python
# To handle the errors in the ship-state by using replace function
df.replace({'NL':'NAGALAND','PB':'PUNJAB','AR':'ARUNACHAL PRADESH','PUNJAB/MOHALI/ZIRAKPUR':'PUNJAB','RJ':'RAJASTHAN', 'PONDICHERRY':'PUDUCHERRY','RAJSHTHAN':'RAJASTHAN'},inplace=True)
```

```python
# Formatting the date column to datetime datatype
#using lambda function to convert Date to string 

df['Date']= pd.to_datetime(df['Date'])
```

```python
# Lets rename teh column Qty ti Quantity
df.rename(columns = {'Qty':'Quantity'}, inplace=True)
df
```

```python
Total_nos_category=df['Category'].value_counts()
Total_nos_category
```

```python
# Lets Begin our Visualisations
# Lets plot our first graph, Category vs quantity

category_counts = Total_nos_category.reset_index()
category_counts.columns = ['Category', 'Occurence Count']



# Create bar plot

plt.figure(figsize=(12, 7))
ax = sns.barplot(x='Category', y='Occurence Count', data=category_counts)
for p in ax.containers:
    ax.bar_label(p, fontsize=14, color='black', padding=2)
plt.title("Products vs Occurence ", fontsize=30)
plt.xlabel("Products", fontsize=25)
plt.ylabel("Occurence Count", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
Sales_per_product=df.groupby(['Category'])['Amount'].sum()
Sales_per_product

Sales = Sales_per_product.reset_index()
Sales.columns = ['Category', 'Amount']
Sales
```

```python
# Create bar plot

plt.figure(figsize=(12, 7))
bx=sns.barplot(x='Amount', y='Category', data=Sales)
for p in bx.containers:
    bx.bar_label(p, fontsize=14, color='black', padding=2)
plt.title("Total Sales By Product",fontsize=30)
plt.xlabel("Sales per product(in 5 millions)", fontsize=20)
plt.ylabel("Products", fontsize=20)
plt.xlim((Sales['Amount'].min(),Sales['Amount'].max()+5000000))
plt.xticks(fontsize=20, rotation=0)
plt.yticks(fontsize=20, rotation=0)
plt.show()
```

```python
#Adding a month Feature for monthly analysis 
df['Month Name'] = df['Date'].dt.strftime('%B')
df['Month Name']
```

```python
# Tracking the monthly sales for top 5 states 
monthly_sales = df.groupby([ 'ship-state','Month Name'])['Amount'].sum().unstack().sort_values('June',ascending=False)
top_5_states=monthly_sales[0:5]
```

```python
top_5_states.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title("Five states with the highest monthly sales")
plt.xlabel("States")
plt.ylabel("Sales(In Millions) ")
plt.xticks(rotation=0)
plt.legend(title='Month Name')
plt.show()
```

```python
#Heat map for a quick glance at count of different products as per the sizes
categorical_sizes = df.groupby(['Size', 'Category'])['Month Name'].count().unstack()


plt.figure(figsize=(12, 6))
sns.heatmap(categorical_sizes, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Product Count Across Different Sizes")
plt.xlabel("Products")  
plt.ylabel("Sizes")
plt.show()
```
